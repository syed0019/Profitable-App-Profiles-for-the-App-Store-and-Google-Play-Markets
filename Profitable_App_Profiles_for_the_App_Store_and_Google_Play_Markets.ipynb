```python
#checking length and names of duplicate apps in android:

duplicate_list = []
unique_list = []

#google play store dataset
for app in android:
    name = app[0]
    if name in unique_list:
        duplicate_list.append(name)
    else:
        unique_list.append(name)

print('Number of Duplicate apps:', len(duplicate_list))
print('\n')
print('Names of Duplicate apps:', duplicate_list[:13])
```

We have examined that `android` dataset has 1181 duplicate names of apps and therefore, we would remove the duplicates for reliability. Before doing so, we need to determine criterion to remove duplicates instead of doing it randomly, as dataset contains necessary information required for analysis.

In below cell, we printed & examined two cells, the main difference happens on the fourth position of each row, which corresponds to the number of reviews. The different numbers show that the data was collected at different times. We can use this to build a criterion for keeping rows. Therefore, we will create a dictionary where each key is a unique app and value will be maximum number of reviews.

```python
# reviewing the data before determining the criterion
print(android_header)
print('\n')
explore_data(android, 0, 2)
```

```python
# creating a dictionary for the removal of duplicates
reviews_max = {}
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if (name in reviews_max) and (reviews_max[name] < n_reviews):
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
reviews_max
```

```python
expected_rows = len(android) - len(duplicate_list)
```

```python
print(len(reviews_max))
print(len(reviews_max) == expected_rows)
```

We will use above created dictionary to remove duplicates by initiating two empty lists, `android_clean` that will store the new/cleaned apps and `already_added` that will just store the apps. Below, we loop through the android data set, and for every iteration:
- We isolate the name of the app and the number of reviews.
- We add the current row (app) to the android_clean list, and the app name (name) to the already_cleaned list, if:
  - The number of reviews of the current app matches the number of reviews of that app as described in the reviews_max dictionary; and
  - The name of the app is not already in the already_added list. We need to add this additional condition to account for those cases where the highest number of reviews of a duplicate app is the same for more than one entry.

```python
#initiate two empty lists
android_clean = [] 
already_added = [] 

for app in android:
    name = app[0]
    n_reviews = float(app[3])

    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

```python
# verifying if dataset `android_clean` has 9659 rows
explore_data(android_clean, 0, 3, True)
```

We have extracted data of non-english apps from both datasets. Now, we will create a function to determine if character(s) in an app is in ASCII and that each character's corresponding encoding number (in ASCII) is between 0 and 127 as this will help as extracting non-ascii characters. In our function below, we will use built-in `ord()` function to find out the corresponding encoding number of each character.

```python
def is_english(a_string):
    for character in a_string:
        if ord(character) > 127:
            return False
    return True
```

The function seems to work fine, but some English app names use emojis or other symbols (â„¢, â€” (em dash), â€“ (en dash), etc.) that fall outside of the ASCII range. Because of this, we'll remove useful apps if we use the function in its current form.

```python
# to minimize loss, we will redefine function to include atmost 3 ASCII characters in an app name
def is_english(a_string):
    non_ascii_charac = 0
    for character in a_string:
        if ord(character) > 127:
            non_ascii_charac += 1
    if non_ascii_charac > 3:
        return False
    return True
```

```python
print(is_english('Docs To Goâ„¢ Free Office Suite'))
print(is_english('Instachat ðŸ˜œ'))
```

After examining the above datasets, we will build a frequency table for the `prime_genre column` of the App Store dataset, and the `Genres` and `Category` columns of the Google Play dataset. To achieve this, We'll build two functions which we can use to analyze the frequency tables:
- One function to generate frequency tables that show percentages
- Another function that we can use to display the percentages in a descending order

```python
def freq_table(dataset, index):
    f_table = {}
    f_total = 0
    
    for genre in dataset:
        f_total += 1
        value = genre[index]
        if value in f_table:
            f_table[value] += 1
        else:
            f_table[value] = 1
            
    f_table_percentages = {}
    for key in f_table:
        percentage = (f_table[key] / f_total) * 100
        f_table_percentages[key] = percentage
    return f_table_percentages
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    sorted_table = sorted(table_display, reverse=True)
    for entry in sorted_table:
        print(entry[1], ':', entry[0])
```

```python
# examining frequency table for `prime genre` column of apple store dataset
display_table(ios_free_apps, -5)
```

#### Most Popular Apps by Genre on the App Store

One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. For the Google Play data set, we can find this information in the Installs column, but for the App Store data set this information is missing. As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the rating_count_tot app.

```python
# Below, we calculate the average number of user ratings per app genre on the App Store:

unique_ios_genre = freq_table(ios_free_apps, -5)

for genre in unique_ios_genre:
    total = 0
    len_genre = 0
    for app in ios_free_apps:
        genre_app = app[-5]
        if genre_app == genre:
            n_user_ratings = float(app[5])
            total += n_user_ratings
            len_genre += 1
    avg_number_ratings = total / len_genre
    print(genre, ':', avg_number_ratings)
```

On average, navigation apps have the highest number of user ratings.

#### Most Popular Apps by Genre on Google Play

For the Google Play market, we actually have data about the number of installs, so we should be able to get a clearer picture about genre popularity.

```python
display_table(android_free_apps, 5)
```

One issue with the above data is that it is not precise. For instance, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, here we don't need very precise data for our purposes â€” we only want to get an idea which app genres attract the most users. 
To perform computations, however, we'll need to convert each install number to float â€” this means that we need to remove the commas and the plus characters.

```python
unique_android_categories = freq_table(android_free_apps, 1)

for category in unique_android_categories:
    total = 0
    len_category = 0
    for app in android_free_apps:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```

On average, communication apps have the most installs: 38,456,119. This number is heavily skewed up by a few apps that have over one billion installs (WhatsApp, Facebook Messenger, Skype, Google Chrome, Gmail, and Hangouts), and a few others with over 100 and 500 million installs:
